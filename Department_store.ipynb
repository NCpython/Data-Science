```python
import pandas as pd
import numpy as np
import sklearn as sk
import sklearn.preprocessing
```

```python
##Feature engineering
##creaing one more column storing the maximum first_visit value
Dept_data['max_first_visit']=Dept_data['first_visit'].max()
Dept_data.head()
```

```python
##converting the columns to datetime type to subtract
Dept_data['max_first_visit'] = pd.to_datetime(Dept_data['max_first_visit'])
```

```python
##storing the difference of max_visit and first_visit
Dept_data['cust_retail_age']=Dept_data['max_first_visit']-Dept_data['first_visit']
```

```python
Dept_data.head()
```

```python
##Dropping two columns that are not required first_visit and max_first-visit
Dept_data.drop(columns=['max_first_visit'],inplace=True)
```

```python
##Creating two different data set
Dept_data1=Dept_data[Dept_data['targeted']==1]
Dept_data2=Dept_data[Dept_data['targeted']==0]
```

```python
len(Dept_data2)
```

```python
Dept_data1.reset_index(inplace = True)
```

```python
Dept_data1.head()
```

```python
Dept_data2.reset_index(inplace = True)
Dept_data2.head()
```

```python
Dept_data2.head()
```

```python
## one hot encoding for 
dummies=pd.get_dummies(Dept_data1['mode_of_payment'])
Dept_data1=pd.concat([Dept_data1,dummies],axis=1)
Dept_data1.drop(columns=['mode_of_payment'],inplace=True)
```

```python
Dept_data1.drop(columns=['index'],inplace=True)
```

```python
Dept_data1['tdColumn']=pd.to_numeric(Dept_data1['cust_retail_age'].dt.days, downcast='integer')
```

```python
Dept_data1.tail()
```

```python
Dept_data1.drop(columns=['cust_retail_age'],inplace=True)
```

```python
X=Dept_data1.iloc[:,3:16]
X
```

```python
Y=Dept_data1.iloc[:,2]
Y
```

```python
##Splitting the data into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
X.dtypes
```

```python
import warnings
warnings.filterwarnings("ignore")
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf1=RandomForestClassifier(n_estimators=20,oob_score=True,class_weight={0:1,1:3},max_features=4,max_depth=6)##,max_leaf_nodes=20,##random_state=None,
                           ##min_samples_leaf=200,min_impurity_decrease=0.05)##,min_impurity_decrease=0.0
                          ##,min_impurity_split=None,min_samples_leaf=4,min_samples_split=55)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf1.fit(X_train,y_train)
print("oob_score:",1- clf1.oob_score_)
y_pred=clf1.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
y_pred_new1 = clf1.predict_proba(X_test)
y_pred_new1=pd.DataFrame(y_pred_new1)
y_pred_new1.columns=['prob_of_o','prob_of_1']
y_pred_new1['prob_cutoff']=np.where(y_pred_new1.prob_of_1>0.7,1,0)
y_pred_new1=y_pred_new1.prob_cutoff
```

```python
##confusion matrix
from sklearn.metrics import classification_report
print (classification_report(y_test, y_pred))
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```
